## Scatterplot

```python
import seaborn as sns
sns.set(style = "white")

mpg = sns.load_dataset("mpg")
```

```python
mpg.head()
```

```python
sns.scatterplot(x="horsepower",y="mpg",data=mpg)
```

```python
sns.set(style="whitegrid")
sns.scatterplot(x="horsepower",y="mpg",data=mpg)
```

```python
sns.scatterplot(x="horsepower",y="mpg",data=mpg,hue='origin').set_title("ScatterPlot")
```

```python
sns.scatterplot(x="horsepower",y="mpg",data=mpg,hue='origin',size='weight').set_title("ScatterPlot")
```

```python
sns.scatterplot(x="horsepower",y="mpg",data=mpg,hue='origin',size='weight',palette='muted').set_title("ScatterPlot")
```
