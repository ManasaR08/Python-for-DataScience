## Scatterplot

```python
import seaborn as sns
sns.set(style = "white")

mpg = sns.load_dataset("mpg")
```

```python
mpg.head()
```

```python
sns.scatterplot(x="horsepower",y="mpg",data=mpg)
```

```python
sns.set(style="whitegrid")
sns.scatterplot(x="horsepower",y="mpg",data=mpg)
```

```python
sns.scatterplot(x="horsepower",y="mpg",data=mpg,hue='origin').set_title("ScatterPlot")
```

```python
sns.scatterplot(x="horsepower",y="mpg",data=mpg,hue='origin',size='weight').set_title("ScatterPlot")
```

```python
sns.scatterplot(x="horsepower",y="mpg",data=mpg,hue='origin',size='weight',palette='muted').set_title("ScatterPlot")
```

```python
import matplotlib.pyplot as plt
plt.style.use("dark_background")
sns.scatterplot(x="horsepower",y="mpg",data=mpg,hue='origin',size='weight',palette='muted').set_title("ScatterPlot")
```

## FacetGrid

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.set(style='ticks',color_codes=True)
sns.FacetGrid(tips,col='time',row='smoker').map(plt.hist,"total_bill")
```

```python
sns.set(style='ticks',color_codes=True)
sns.FacetGrid(tips,col='sex',row='smoker').map(plt.hist,"total_bill")
```

```python
sns.set(style='ticks',color_codes=True)
sns.FacetGrid(tips,col='time',row='smoker').map(plt.scatter,"total_bill","tip",edgecolor='w')
```

```python
sns.set(style='ticks',color_codes=True)
sns.FacetGrid(tips,col='time',hue='smoker').map(plt.scatter,"total_bill","tip",edgecolor='w').add_legend()
```

```python
import numpy as np
bins = np.arange(0,65,3)
sns.FacetGrid(tips,col="day",height=4).map(plt.hist,"total_bill",bins=bins)
```

```python
att = sns.load_dataset("attention")
att.head()
```

```python
sns.FacetGrid(att,col="subject").map(plt.plot,"solutions","score",marker=".")
```

```python
sns.FacetGrid(att,col="subject",col_wrap=5).map(plt.plot,"solutions","score",marker=".")
```
